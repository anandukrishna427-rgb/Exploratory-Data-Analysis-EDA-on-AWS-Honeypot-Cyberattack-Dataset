# Data Preparation

##### Converting Datetime Column to Proper Datetime Format

```python
df['datetime']=pd.to_datetime(df['datetime'])
```

##### Removing Unnecessary or Empty Columns from the Dataset

```python
df.drop("Unnamed: 15",axis=1,inplace=True)
```

```python
df.drop("type",axis=1,inplace=True)
```

```python
df.drop("src",axis=1,inplace=True)
```

```python
df.drop("postalcode",axis=1,inplace=True)
```

```python
df.drop("cc",axis=1,inplace=True)
```

```python
df.drop('localeabbr',axis=1,inplace=True)
```

```python
df=df.dropna()
```

##### Renaming Columns for Better Readability and Clarity


```python
df=df.rename(columns={"datetime":'Datetime','host':'Honeypot Host','proto':'Protocol','spt':'Source Port','dpt':'Destination Port',
                   'srcstr':'Source IP','country':'Country','locale':'City','latitude':'Latitude','longitude':'Longitude'})
```

##### Resetting DataFrame Index After Cleaning


```python
df.reset_index(drop=True,inplace=True)
```

##### Extracting Date and Time Components from Datetime Column

```python
df['Year']=df['Datetime'].dt.year
df['Month']=df['Datetime'].dt.month_name()
df['Day']=df['Datetime'].dt.day_name()
df['Hour']=df['Datetime'].dt.hour
df['Minute']=df['Datetime'].dt.minute
```

# Outlier Analysis

##### Source Port Outlier Detection and Analysis

```python
Q1 = df['Source Port'].quantile(0.25)
Q3 = df['Source Port'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Source Port'] < lower) | (df['Source Port'] > upper)]
print(f"Found {len(outliers)} outlier source ports")
print(outliers[['Source Port', 'Source IP', 'Destination Port', 'Protocol']].head())
print("jdfhjhahfjs")
```

##### Identification of High-Priority Attack Indicators

```python
df['is_low_source_port'] = df['Source Port'] < 1024
df['is_repeated_port_ip'] = df.groupby(['Source IP', 'Source Port'])['Source Port'].transform('count') > 50
rare_dest_ports = df['Destination Port'].value_counts().tail(50).index
df['is_rare_dest_port'] = df['Destination Port'].isin(rare_dest_ports)

high_priority = df[
    df['is_low_source_port'] |
    df['is_repeated_port_ip'] |
    df['is_rare_dest_port']
].copy()

print(f"High-priority outliers: {len(high_priority)}")
print("\nTop 10 High-Priority Attacks:")
print(high_priority[['Datetime', 'Source IP', 'Source Port', 'Destination Port', 'Protocol', 'Country']]
      .head(10).to_string(index=False))
```

##### Detection and Summary of Suspicious Attack Patterns

```python
low_port_attacks = df[df['Source Port'] < 1024]
rare_port_attacks = df[df['Destination Port'].isin(rare_dest_ports)]

suspicious_attacks = df[
    (df['Source Port'] < 1024) |
    (df['Destination Port'].isin(rare_dest_ports))
].copy()

print(f"Total suspicious attacks: {len(suspicious_attacks)}")
print("\nTop 10 High-Interest Attacks:")
print(suspicious_attacks[['Datetime', 'Source IP', 'Source Port', 'Destination Port', 'Protocol', 'Country']]
      .head(10).to_string(index=False))
```

##### Temporal Analysis of Top Attacker IP Activity

```python
top_ips = suspicious_attacks['Source IP'].value_counts().head(5).index
df_suspicious_ips = suspicious_attacks[suspicious_attacks['Source IP'].isin(top_ips)]

sns.boxplot(x='Source IP', y='Hour', data=df_suspicious_ips)
plt.title("When Top Hackers Attack (Suspicious Attacks)")
plt.xlabel("Hacker IP")
plt.ylabel("Time of Day (0–23)")
plt.xticks(rotation=45)
plt.show()
```

##### Temporal Distribution of Suspicious Attacks Across Honeypot Locations

```python
sns.boxplot(x='Honeypot Host', y='Hour', data=suspicious_attacks)
plt.title("When Suspicious Attacks Happen by Honeypot Location")
plt.xlabel("Honeypot Location")
plt.ylabel("Time of Day (0–23)")
plt.xticks(rotation=45)
plt.show()
```

##### Geographic Analysis of Attack Origins by Country

```python
top_countries = suspicious_attacks['Country'].value_counts().head(5).index
df_suspicious_top = suspicious_attacks[suspicious_attacks['Country'].isin(top_countries)]

sns.boxplot(x='Country', y='Latitude', data=df_suspicious_top)
plt.title("Where Suspicious Attacks Come From by Country")
plt.xlabel("Country")
plt.ylabel("Location (Latitude)")
plt.xticks(rotation=45)
plt.show()
```

##### Frequency Analysis of Cross-Country Attack Patterns (Top 7 Attacking Countries)

```python
host_to_country = {
    'groucho-oregon': 'United States',
    'groucho-us-east': 'United States',
    'groucho-tokyo': 'Japan',
}
suspicious_attacks['Target Country'] = suspicious_attacks['Honeypot Host'].map(host_to_country)

top_countries = suspicious_attacks['Country'].value_counts().head(7).index
df_suspicious_top = suspicious_attacks[suspicious_attacks['Country'].isin(top_countries)]

attack_counts = df_suspicious_top.groupby(['Country', 'Target Country']).size().unstack(fill_value=0)

plt.figure(figsize=(12, 8)) 
attack_counts.plot(kind='bar', stacked=False, width=0.8, colormap='tab10')
plt.title("How Often Countries Attack Other Countries (Top 7)")
plt.xlabel("Attacking Country")
plt.ylabel("Number of Attacks")
plt.legend(title="Attacked Country", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
